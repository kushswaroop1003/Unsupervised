```python
movies_rating_count = data_movie.groupby("title")["rating"].count().reset_index().rename(columns = {"rating":"TotalRatingCOunt"})
```

```python
data_movie= data_movie.merge(movies_rating_count, on = "title")
```

```python
data_movie = data_movie[["userId","title","rating","TotalRatingCOunt"]]
```

```python
data_movie.head(2)
```

```python
movies_rating_count["TotalRatingCOunt"].describe()
```

```python
popularity_threshold = 50
popular_movies= data_movie.query("TotalRatingCOunt >= @popularity_threshold")
popular_movies.head()
```

```python
#Create a Pivot Table
```

```python
movie_features=popular_movies.pivot_table(index='title',columns='userId',values='rating').fillna(0)
movie_features.head()
```

```python
from scipy.sparse import csr_matrix

movie_features_matrix = csr_matrix(movie_features.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_matrix)
```

```python
query_index = np.random.choice(movie_features.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
movie_features.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features.index[indices.flatten()[i]], distances.flatten()[i]))
```
